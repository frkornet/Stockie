```python
stocks['RT']  = 1
stocks['BLK'] = 1

# set the unique Renaissance Technoclogies stocks to zero for BlackRock indicator field
for cusip in (stocks_cusip - blk_data_cusip):
    idx = stocks.CUSIP == cusip
    stocks.loc[idx, 'BLK'] = 0

sum(stocks.BLK == 1), sum(stocks.RT == 1), sum(stocks.RT == 1) - sum(stocks.BLK == 1)
```

```python
stocks.info()
```

```python
unique_blk_cusip = blk_data_cusip - stocks_cusip
cols = stocks.columns
unique_blk_data = pd.DataFrame(columns=cols)

for cusip in unique_blk_cusip:
    row = blk_data [ blk_data.CUSIP == cusip]
    assert len(row) == 1, "Cannot have multiple rows per CUSIP!"
    
    idx = row.index
    row.loc[idx, 'TICKER'] = np.NaN
    row.loc[idx, 'RT']     = 0
    row.loc[idx, 'BLK']    = 1

    unique_blk_data = unique_blk_data.append(row)
    
unique_blk_data.RT = unique_blk_data.RT.astype(int)
unique_blk_data.BLK = unique_blk_data.BLK.astype(int)
unique_blk_data.head(10)
```
